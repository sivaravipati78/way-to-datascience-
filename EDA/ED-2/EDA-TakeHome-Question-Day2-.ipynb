####  1.7 importing import skew, norm

```python
from scipy.stats import skew,norm
```

#### 1.8 Create the numeric_features to remove skewness

```python
numfeat = df1.columns[:-1]
numfeat
```

#### 1.9 Create the skewed_features = wines[numeric_features]

```python
skewed_features =df1[numfeat]
skewed_features.head()
```

#### 1.10 Create DataFrame for skewness?

```python
skewness=pd.DataFrame(skewed_features.skew(),columns=["skewness_of_parameters"])
skewness
```

#### 1.11 Create to Get only higest skewed = [abs(skewness) > 0.7] features?

```python
skewness[abs(skewness["skewness_of_parameters"])>.7]
```

```python
# Get only higest skewed features
skewness[skewness["skewness_of_parameters"]==skewness["skewness_of_parameters"].max()]
```

#### 1.12 Create skewness.index and new skewed_features2

```python
#Create Empty list 
skewed_features2=[]
#create for loop for skewness.index
for i in skewness.index:
    skewed_features2.append(skew(df1[i]))
#create skewed_features2 for skewness.index
```

#### 1.13 Create new DataFrame for skewed_features2?

```python
#compute skewness

skewness_features2=pd.DataFrame(skewed_features2,index=skewness.index,columns=["skewness_of_parameters2"])
# concat both skewness, skewness2
skewness_features2
```

```python
a=pd.concat([skewness,skewness_features2],axis=1)
a
```
