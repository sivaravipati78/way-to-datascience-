```python
df.Value.replace(to_replace=r'\A€',value ='', regex=True,inplace=True)
```

```python
#https://docs.python.org/2.0/ref/strings.html
#https://pandas.pydata.org/pandas-docs/version/0.21/api.html#series
df.Value =(df.Value.replace(r'[KM]+$', '', regex=True).astype(float) * \
df.Value.str.extract(r'([KM])',expand=False).fillna(2410695.8861976163).replace(['K','M'], [10**3, 10**6]).astype(int))

#df1.Value = (df1.Value.replace(r'[KM]+$', '', regex=True).astype(float) * \
#df1.Value.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(2410695.8861976163).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
df.Value.dtype
```

```python
df.Wage.replace(to_replace=r'^€',value ='', regex=True,inplace=True)
```

```python
df.Wage =(df.Wage.replace(r'[KM]','',regex=True ).astype(float) * \
df.Wage.str.extract(r'([KM])',expand=False).fillna(2410695.8861976163).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
df.Joined = df.Joined.str.split(',').str[1].fillna(method = 'ffill' ).astype(int)
```

```python
df.Joined = df.Joined.astype(int)
```

```python
df['Contract Valid Until'].fillna(method ='ffill',inplace=True)
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
```

```python
df.Height.replace(  =r"'",value ='.', regex=True,inplace =True )                 
```

```python
df.Height =df.Height.fillna(method ='ffill').astype(float)
```

```python
df.Weight.replace(to_replace=r'lbs$', value='', inplace=True,regex=True)
```

```python
df.Weight =df.Weight.astype(float)
```

```python
df['Release Clause'].replace(to_replace=r'^€', value='', inplace=True,regex=True)
```

```python

df['Release Clause'] =(df['Release Clause'].replace(r'[KM]+$', '', regex=True).astype(float) * \
df['Release Clause'].str.extract(r'([KM])',expand=False).fillna(2410695.8861976163).replace(['K','M'], [10**3, 10**6]).astype(int))
```
