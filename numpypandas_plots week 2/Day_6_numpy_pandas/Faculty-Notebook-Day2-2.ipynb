* DataFrame is a tabular data structure in which data is laid out in rows and column format (similar to a CSV or tables in SQL file), but it can also be used for higher dimensional data sets. 

* The DataFrame object can contain homogenous and heterogenous values, and can be thought of as a logical extension of Series data structures. <span style="font-family: Arial; font-weight:bold;font-size:1.50em;color:#7604ce;"> Dataframe using Dictionaries

The default manner to create a DataFrame from python is to use a list of dictionaries. In this case each dictionary key is used for the column headings. A default index will be created automatically:

```python
Image('7.png')
```

```python
GL = ({"Location":"Bangalore","Jan":150,"Feb":80},
     {"Location":"Chennai","Jan":200,"Feb":70})
Df = pd.DataFrame(GL)
Df
```

```python
GL = ({"Location":["Bangalore","Chennai"],
       "Jan":[150,80],
       "Feb":[200,70]})

df = pd.DataFrame.from_dict(GL)
df
```

<span style="font-family: Arial; font-weight:bold;font-size:1.50em;color:#7604ce;"> DataFrame using List
    
The other option for creating your DataFrames from python is to include the data in a list structure.

The first approach is to use a row oriented approach using pandas from_records . This approach is similar to the dictionary approach but you need to explicitly call out the column labels.

```python
Image('8.png')
```

```python
GL = [("Bangalore",150,80),
      ("Chennai",200,70)]
Labels = ["Location","Jan","Feb"]
df = pd.DataFrame.from_records(GL,columns=Labels)
df
```

```python
GL = [("Location", ["Bangalore","Chennai"]),
       ("Jan",[150,200]),
       ("Feb",[80,70])]

df = pd.DataFrame.from_items(GL)
df
```

<span style="font-family: Arial; font-weight:bold;font-size:1.50em;color:#7604ce;">  Indexing with iloc:
    
**.iloc [1:m, 1:n] **  is used to select or index rows based on their position from 1 to m rows and 1 to n columns

<span style="font-family: Arial; font-weight:bold;font-size:1.50em;color:#7604ce;"> Indexing a DataFrame using .loc[ ] :
    
This function selects data by the label of the rows and columns. The df.loc indexer selects data in a different way than just the indexing operator. It can select subsets of rows or columns. It can also simultaneously select subsets of rows and columns.

```python
Image('9.png')
```

#### Select row by using row number in pandas with .iloc

```python
# select first 2 rows
df.iloc[:2]
# or
df.iloc[:2,]
```

```python
# select 3rd to 5th rows
df.iloc[2:5]
# or 
df.iloc[2:5,]
```

#### Select column by using column number in pandas with .iloc

```python
# select first 2 columns
df.iloc[:,:2]
```

```python
# select 1st and 4thcolumn
df.iloc[:,[0,3]]
```

#### Select values by using .iloc
#### Select 2nd row and 3rd column value

```python
# Select 2nd row and 3rd column value
df.iloc[1,2]
```

```python
Data= {'Day':[1,2,3,4,5,6], "Students":[1000, 700,6000,1000,400,350], "Refer_Students":[20,20, 23,15,10,34],"python_student":[80,90,70,60,70,20]}
 
data= pd.DataFrame(Data)
 
print(data)
```

### Creating a rank column and passing the returned rank series 

```python
data["Rank"] = data["Students"].rank() 
data   
```

### sorting with respect to Students column

### Creating a rank column and passing the returned rank series 

```python
data["Rank"] = data["python_student"].rank(method ='average')  
data 
```

#### `Ranking` the dataframe in python pandas on `ascending order`:

```python
# Ranking of score ascending order
 
data['Rank']=data['python_student'].rank(ascending=1)
data
```

#### `Ranking` the dataframe in python pandas on `descending order`.

```python
 
data['Rank']=data['python_student'].rank(ascending=0)
data
```

#### `Rank` the dataframe in python pandas by `minimum value` of the rank.

```python
data['Rank']=data['python_student'].rank(ascending=0,method='min')
data
```

#### Rank the dataframe in python pandas by `dense rank`.

```python
data['Rank']=data['python_student'].rank(ascending=0,method='dense')
data
```
