# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sn
%matplotlib inline 
# jupyter command  # to show the output image
```

```python
x1 = np.array([2,4,6,8,10])
y1 = x1*2
#plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
x2 =x1
y2 = x2**2

plt.plot(x1,y1,linestyle ='solid',label ='Plot1',marker='o',color ='r')
plt.plot(x2,y2,linestyle ='dotted',label ='Plot2',marker='*',color ='g')

#plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)

plt.ylabel('Y-Axis')
plt.xlabel('x-Axis')

plt.xticks([2,4,6,8,10])
plt.yticks([12,24,36,48,90])

plt.title('First plot')
plt.legend(fontsize =18,loc =0)# giving system to fix the default valueloc =0
plt.show()
print()
```

```python


stock_names=['GOOGLE','APPLE','IBM','INTEL','CISCO']
stock_values=[1000,800,600,400,200]

x_pos=list(range(len(stock_names)))
x_pos[0]=4 # for getting overlapped on same x tick.
plt.style.use('ggplot')
plt.bar(x_pos,stock_values,width=0.35,align='center',color='rgb',alpha=0.8)
plt.bar(x_pos,stock_values,width=0.35,align='center',color='k',alpha=0.8,bottom =stock_values )

plt.xticks(x_pos,stock_names)
plt.ylabel('Price')
plt.xlabel('Name')
plt.title('Stock Price')
         
plt.show()
```

```python
stock_names=['GOOGLE','APPLE','IBM','INTEL','CISCO']
stock_values=[1000,800,600,400,200]
stock_rev =[200,180,350,120,750]
bar_width =0.35
x_pos=np.arange(len(stock_names)) # we cant use list as list cant be added to values with out iteration
x_pos[0]=4 # for getting overlapped on same x tick.
plt.style.use('ggplot')
plt.bar(x_pos,stock_values,width=0.35,align='center',color='r',alpha=0.8)
plt.bar(x_pos+bar_width ,stock_rev,width=0.35,align='center',color='g',alpha=0.8)

plt.xticks(x_pos,stock_names)
plt.ylabel('Price')
plt.xlabel('Name')
plt.title('Stock Price')
         
plt.show()
```

```python
stock_names=['GOOGLE','APPLE','IBM','INTEL','CISCO']
stock_values=[1000,800,600,400,200]

y_pos=list(range(len(stock_names)))
y_pos[0]=4 # for getting overlapped on same x tick.
#plt.style.use('ggplot')
#plt.style.use('dark_background')
#plt.style.use('bmh')

#plt.style.use('seaborn-poster')
plt.style.use('classic')
plt.barh(y_pos,stock_values,height=0.4,align='center',color='rgb',alpha=0.8)
plt.yticks(y_pos,stock_names)
plt.ylabel('Price')
plt.xlabel('Name')
plt.title('Stock Price')
         
plt.show()
```

# pie chart

```python
slices =[46,10,20,14]
activities =['Agricultue','Transpor','IT','Banking']
colors =['c','m','r','b']
plt.style.use('ggplot')
plt.pie(labels =activities,colors = colors,x = slices,explode=(0,0,0,0.3),autopct='%.1f',center=(0, 0),frame=True,shadow=True)
plt.legend(activities)
plt.title('Pie chart')
plt.show()
```
