9.	Convert Amount.Requested', 'Interest.Rate','Debt.To.Income.Ratio','Open.CREDIT.Lines','Revolving.CREDIT.Balance to Numeric.

```python
for col in ['Amount.Requested', 'Interest.Rate','Debt.To.Income.Ratio',
            'Open.CREDIT.Lines','Revolving.CREDIT.Balance']:
    total[col]=pd.to_numeric(total[col],errors='coerce')
    
```

11.	Clean the employment Length Varaible.

14.	Create dummy variables for Loan Length,Loan purpose, Home ownership and state columns, make sure that there are not too many dummies created, check the frequency and control the dummies based on frequency.

```python
for col in cat_cols:
    print(col)
    freqs=total[col].value_counts()
    print(freqs)
    k=freqs.index[freqs>20][:]
    print(k)
    for cat in k:
        print(cat)
        name=col+'_'+cat
        print(name)
        total[name]=(total[col]==cat).astype(int)
    del total[col]
    print(col)
```

15.	Check the missing values and impute them with mean

```python
total.isna().sum()
```

```python
total.head()
```

```python
for col in total.columns:
    if (col not in ['Interest.Rate','data'])& (total[col].isnull().sum()>0):
        total.loc[total[col].isnull(),col]=total.loc[total['data']=='train',col].mean()
```

16.	Separate the train and test datasets

```python
total.head()
```

splitting back to the train and test data

```python
Trainprocess=total[total['data']=='train']
del Trainprocess['data']
Testprocess=total[total['data']=='test']
Testprocess.drop(['Interest.Rate','data'],axis=1,inplace=True)
```

```python
Trainprocess.shape,Testprocess.shape
```

converting the data to a standard scalar scale

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
traintscalar =scaler.fit_transform(Trainprocess)
scaledtrain = pd.DataFrame(traintscalar,columns =Trainprocess.columns)
scaledtrain.shape
```

```python
testtscalar =scaler.fit_transform(Testprocess)
scaledtest = pd.DataFrame(testtscalar,columns =Testprocess.columns)
scaledtest.shape
```

```python
scaledtrain.shape,scaledtest.shape
```

splitting train into xtrain ,acv,xtest

```python
from sklearn.model_selection import train_test_split
```

```python
sctrain,scrossval=train_test_split(scaledtrain,test_size=0.2,random_state=2)
```

```python
# Notice that only train data is used for imputing missing values in both train and test 

x_train=sctrain.drop('Interest.Rate',axis=1)
y_train=sctrain['Interest.Rate']
x_cv=scrossval.drop('Interest.Rate',axis=1)
y_cv=scrossval['Interest.Rate']
xtest =scaledtest
```

```python
x_train.shape,y_train.shape,x_cv.shape,y_cv.shape,scaledtest.shape
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lambdas=np.linspace(1,100,100)
```

```python
params={'alpha':lambdas}
```

```python
model=Ridge(fit_intercept=True)
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')
```

```python
grid_search.fit(x_traintotal,y_traintotal)
```

```python
grid_search.best_estimator_
```

```python
grid_search.cv_results_
```

```python
test_pred=grid_search.predict(scaledtest)
```
