#### slice() Parameters
slice() mainly takes three parameters which have the same meaning in both constructs:

1. **start** - starting integer where the slicing of the object starts
2. **stop** - integer until which the slicing takes place. The slicing stops at index stop - 1.
3. **step** - integer value which determines the increment between each index for slicing

### 1. Set Operations
     A. Union 
     B. Intersection
     C. Difference
     D. Symmetric difference

#### 1.A) Union:

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Union :", A | B) 
```

#### 1.B. Intersection

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Intersection :", A & B) 
```

#### 1.C. Difference

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Difference :", A - B) 
```

####  1.D. Symmetric difference:

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Symmetric difference :", A ^ B) 
```

#### 2.C) dict.fromkeys():

```python
# seq is a tuple
seq = ('name', 'age', 'sex')
# tuple is type casted into dict and all the elements are converted into key's. So they key will not have any values.
dict = dict.fromkeys(seq)

print ("New Dictionary : %s" %  str(dict))
# assign a default value as 10 for all the keys
dict = dict.fromkeys(seq, 10)
print ("New Dictionary : %s" %  str(dict))
```

### IF ELIF ELSE

```python
un = "greatlearnin"
pwd = "chennai@12"

if (un == "greatlearning" and pwd == "chennai@123"):
    print("Succefully logged in!!")
elif (un == "greatlearning" and pwd != "chennai@123"):
    print("User name is right, but not the password!")
elif (un != "greatlearning" and pwd == "chennai@123"):
    print("User name is not right, but the password is right!")
else:
    print("Both the username and password is not right")
```
